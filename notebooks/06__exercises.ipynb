```python
# Sample of original dataset
cidadesDigitais.head()
```

```python
# Define the column STATUS as index of dataframe
cidadesUF_3 = cidadesDigitais.set_index('STATUS'); cidadesUF_3
```

```python
# Aplicando Group By with hierarquival
cidadesUF_3 = cidadesUF_3.groupby(level='STATUS')
```

```python
cidadesUF_3.head()
```

```python
# Grouping by UF and STATUS
cidadesUF_4 = cidadesDigitais.set_index(['UF','STATUS'])
cidadesUF_4 = cidadesUF_4.groupby(level=['UF','STATUS'])
cidadesUF_4.sum()
```

```python
# Aggregating with pd.aggregate() and numpy
cidadesUF_4.aggregate([np.sum,np.mean,np.size])
```
